# The complex exponential $e^{st}$


```python
t = np.arange(0,7,0.01)
ω = 10
σ = 0

s = σ + 1j*ω

f = np.exp(s*t)

plt.figure()
plt.plot(t, f.real, label ='$\Re[f(t)]$')
plt.plot(t, f.imag, label ='$\Im[f(t)]$')
plt.xlim(min(t), max(t))
plt.grid()
plt.xlabel('tempo (s)')
plt.legend()

plt.figure()
plt.plot(f.real,f.imag, label ='$f(t)$')
plt.grid()
plt.xlabel('real')
plt.ylabel('imag')
plt.legend()
```

```python
figName = './figures/complexExponential_sigma0.gif'

genGIF(f.real, f.imag, figName, xlabel='Real', ylabel='Imag.', 
       title = '$f(t)=e^{st}$,' + f's = σ + jω, σ = {σ}, ω = {ω}', squareAxes=True,
       fram=200, inter=80)

Image(figName, width=500)
```

```python
figName = './figures/realPart_cplxExp_sigma0.gif'
genGIF(t, f.real, figName, xlabel='Tempo', ylabel='$\Re[e^{st}]$', 
       title = '$f(t)=\Re[e^{st}]$,' + f's = σ + jω, σ = {σ}, ω = {ω}', 
       squareAxes = False, fram=200, inter=80)

Image(figName, width=500)
```

```python
figName = './figures/imagPart_cplxExp_sigma0.gif'
genGIF(t, f.imag, figName, xlabel='Tempo', ylabel='$\Im[e^{st}]$', 
       title = '$f(t)=\Im[e^{st}]$,' + f's = σ + jω, σ = {σ}, ω = {ω}', 
       squareAxes=False, fram=200, inter=80)

Image(figName, width=500)
```
