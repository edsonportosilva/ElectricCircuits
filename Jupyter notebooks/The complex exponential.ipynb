```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
from circuit.utils import round_expr, symdisp, symplot, genGIF

# temp workaround
import warnings
from matplotlib import MatplotlibDeprecationWarning
warnings.filterwarnings('ignore', category=MatplotlibDeprecationWarning)

plt.rcParams['figure.figsize'] = 6, 4
plt.rcParams['legend.fontsize'] = 13
plt.rcParams['lines.linewidth'] = 2
plt.rcParams['axes.grid'] = False
```

# The complex exponential $e^{st}$


```python
t = np.arange(0,7,0.01)
ω = 10
σ = -1

s = σ + 1j*ω

f = np.exp(s*t)

plt.figure()
plt.plot(t, f.real, label ='$\Re[f(t)]$')
plt.plot(t, f.imag, label ='$\Im[f(t)]$')
plt.xlim(min(t), max(t))
plt.grid()
plt.xlabel('tempo (s)')
plt.legend()

plt.figure()
plt.plot(f.real,f.imag, label ='$f(t)$')
plt.grid()
plt.xlabel('real')
plt.ylabel('imag')
plt.legend()
```

```python
figName = './figures/complexExponential.gif'
genGIF(f.real, f.imag, figName, xlabel='real', ylabel='imag', title = f's = σ + jω, σ = {σ}, ω = {ω}', fram=200, inter=80)

Image(figName, width=500)
```

```python
figName = './figures/realPart_cplxExp.gif'
genGIF(f.real, t, figName, ylabel='tempo', xlabel='$\Re[e^{st}]$', title = f's = σ + jω, σ = {σ}, ω = {ω}', fram=200, inter=80)

Image(figName, width=500)
```

```python
figName = './figures/imagPart_cplxExp.gif'
genGIF(t, f.imag, figName, xlabel='tempo', ylabel='$\Im[e^{st}]$', title = f's = σ + jω, σ = {σ}, ω = {ω}', fram=200, inter=80)

Image(figName, width=500)
```
