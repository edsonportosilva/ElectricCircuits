```python
# definindo vari√°vel tempo
t = sp.symbols('t', real=True)

i = 5*sp.exp(-100*t)
v = 400*sp.exp(-100*t)

p = v*i

symdisp('i(t) = ', i, 'A')
symdisp('v(t) = ', v, 'V')
symdisp('p(t) = ', p, 'W')
```

```python
intervalo = np.linspace(0, 0.05, 1000)

symplot(t, v, intervalo, funLabel='$v(t) = 400e^{-100t}\; V$', 
        xlabel = 'Tempo [ms]',
        ylabel='Tens√£o [V]', 
        figsize=(2,2),
        xfactor=1e-3,
        yfactor=1)
```

```python
symplot(t, i, intervalo, funLabel='$i(t) = 5e^{-100t}\; A$', 
        xlabel = 'Tempo [ms]',
        ylabel='Corrente [A]', 
        figsize=(2,2),
        xfactor=1e-3)
```

```python
symplot(t, p, intervalo, funLabel='$p(t) = 2000e^{-200t}\; W$', 
        xlabel = 'Tempo [ms]',
        ylabel='Pot√™ncia [W]', 
        figsize=(2,2),
        xfactor=1e-3)
```

```python
symdisp('w_{total} = \int_0^\infty p(t)dt ='+ f'\int_0^\infty {latex(p)}dt =', sp.integrate(p, (t, 0, 1.0)), 'J')
```
