```python
# definindo vari√°vel tempo
t = sp.symbols('t', real=True)

i = 5*sp.exp(-100*t)
v = 400*sp.exp(-100*t)

p = v*i

symdisp('i(t) = ', i, 'A')
symdisp('v(t) = ', v, 'V')
symdisp('p(t) = ', p, 'W')
```

```python
intervalo = np.linspace(0, 0.1, 1000)

symplot(t, v, intervalo, funLabel='$v(t)$')
```

```python
symplot(t, i, intervalo, funLabel='$i(t)$')
```

```python
symplot(t, p, intervalo, funLabel='$p(t)$')
```
