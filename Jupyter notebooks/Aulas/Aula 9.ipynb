## Solução dos Exercícios de fixação 

### Exemplo 1:

```python
t = sp.symbols('t', real=True)
i = sp.Piecewise((0, t<0), (10*t*sp.exp(-5*t), t>=0))

L = 100e-3

v = sp.piecewise_fold(L*sp.diff(i,t).simplify())

symdisp('i_L(t) = ', i, 'A')
symdisp('v_L(t) = ', v, 'V')
```

```python
intervalo = np.arange(-0.5,2,0.001)

symplot(t, [i, v], intervalo, funLabel = ['$i_L(t)$ [A]','$v_L(t)$ [V]'],
        xlabel = 't [s]',        
        figsize=(3,2),        
        yfactor=1)
```

### Exemplo 2:

```python
t = sp.symbols('t', real=True)
v = sp.Piecewise((0, t<0), (20*t*sp.exp(-10*t), t>=0))

L = 100e-3

i = sp.piecewise_fold((1/L)*sp.integrate(v,(t, 0, t))).doit()

symdisp('i_L(t) = ', i, 'A')
symdisp('v_L(t) = ', v, 'V')
```

```python
intervalo = np.arange(-0.5,2,0.001)

symplot(t, [i, v], intervalo, funLabel = ['$i_L(t)$ [A]','$v_L(t)$ [V]'],
        xlabel = 't [s]',        
        figsize=(3,2),        
        yfactor=1)
```
